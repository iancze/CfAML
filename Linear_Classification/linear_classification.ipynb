#Chapter 4: Linear Classification

The goal of linear classification is to take an input vector, $\mathbf{x}$ and assign it to one of K discrete classes, $C_k$ where $k = 1, ..., K$. 
The decision surfaces are linear functions of the input vector, $\mathbf{x}$.
Data sets whose classes can be separated exactly by linear decision surfaces are said to be 'linearly separable'.

#4.1 Discriminant Functions

<h3>4.1.1 Two classes</h3>

A discriminant is a function that takes an input vector ${\bf x}$ and assifgns is to one of $K$ classes, denoted $C_k$. Consider first the case of 2 classes. The simplest representation of a linear discriminant function is obtained by taking a linear function of the input vector so that 

$$y(\mathbf{x}) = \mathbf{w}^T\mathbf{x} + w_0,$$

where $\mathbf(w)$ is the weight vector and $w_0$ is the bias. In this simplest case, an input vector, $\mathbf{x}$ is assigned to class $C_1$ if $y(\mathbf{x}) \geq 0$ and to $C_2$ otherwise. So the decision boundary is defined by $y(\mathbf{x}) = 0$. The vector $\mathbf{w}$ defines the orientation of the decision surface and $w_0$ defines the the location:

$$\mathbf{w}^T(\mathbf{x_A}-\mathbf{x_b}) = 0,$$

$$\frac{\mathbf{w}^T\mathbf{x}}{||\mathbf{w}||} = - \frac{w_0}{||\mathbf{w}||}.$$

<h3>4.1.2 Multiple classes</h3>

In this case, consider a K-class discriminant comprising L linear functions of the form

$$y_k(\mathbf{x}) = \mathbf{w_k}^T\mathbf{x} + w_{k0}$$

A point $\mathbf{x}$ will have a class $C_k$ if $y_k(\mathbf{x}) > y_j(\mathbf{x})$ for all $j \ne k$. 
The decision boundary between class $C_k$ and class $C_j$ is therefore give by $y_k(\mathbf{x}) = y_j(\mathbf{x})$ and hence corresponds to a $(D-1)$ dimensional hyperplane defined by 

$$(\mathbf{w_k} - \mathbf{w_j})^T\mathbf{x} + (w_{k0} - w_{j0}) = 0 $$

<h3>4.1.3 Least squares for classification</h3>

Switch to vector notation:

$$\mathbf{y}(\mathbf{x}) = \tilde{\mathbf{W}}^T\tilde{\mathbf{x}}$$

A new input, $\mathbf{x}$ will be assigned to the class for which $y_k = \mathbf{\tilde{w}}_k^T\mathbf{\tilde{x}}$ is largest. $\mathbf{\tilde{W}}$ is determined by minimising a sum-of-squares error function. Consider a training data set $\{\mathbf{x}_n, \mathbf{t}_n\}$ where $n = 1,...,N$, and define a matrix $\mathbf{T}$ whose $n^{th}$ row is the vector $\mathbf{t}^T_n$ and a matrix $\tilde{\mathbf{X}}$ whose $n^{th}$ row is $\mathbf{x}^T_n$. The sum-of-squares error function can be written

$$E_D(W) = \frac{1}{2}Tr\left\{(\tilde{\mathbf{X}}\tilde{\mathbf{W}} - \mathbf{T})^T(\tilde{\mathbf{X}}\tilde{\mathbf{W}} - \mathbf{T})\right\}$$

Maximum likelihood solution for $\tilde{\mathbf{W}}$:

$$\tilde{\mathbf{W}} = (\tilde{\mathbf{X}}^T\tilde{\mathbf{X}})^{-1}\tilde{\mathbf{X}}^T\mathbf{T},$$

and so the discriminant function can be written

$$\mathbf{y}(\mathbf{x}) = \tilde{\mathbf{W}}^T\tilde{\mathbf{x}} = \mathbf{T}^T\left[(\tilde{\mathbf{X}}^T\tilde{\mathbf{X}})^{-1}\tilde{\mathbf{X}}^T\right]^T\tilde{\mathbf{x}}$$

#Example

```python
M = np.ones((np.shape(X)[0], np.shape(X)[1]+1))
Z = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, y))

M[:, 0] = X[:, 0]
M[:, 1] = X[:, 1]
M[:, 2] = np.ones(200)

Z = np.linalg.solve(np.dot(M.T, M), np.dot(M.T, y))
print Z
```

```python
xs = np.linspace(-2, 7)
#ys = Z[0] - Z[1]*xs 
ys = - Z[2]/Z[1] - (Z[0]/Z[1])*xs

fig = plt.figure(figsize=(5, 3.75))
ax = fig.add_subplot(111)
ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.binary, zorder=2)
ax.plot(xs, ys)
```

<h3>4.1.4 Fisher's Linear Discriminant</h3>

Maximise mean separation, minimise within-class variance, i.e. maximise

$$J(\mathbf{w}) = \frac{(m_1-m_2)^2}{s_1^2 + s_2^2}.$$

or

$$J(\mathbf{w}) = \frac{\mathbf{w}^T\mathbf{S}_b\mathbf{w}}{\mathbf{w}^T\mathbf{S}_w\mathbf{w}} $$

Where $\mathbf{S}_b$ is the between class covariance matrix and $\mathbf{S}_w$ is the within class covariance matrix.
