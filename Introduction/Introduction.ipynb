## Curve fitting as an example of training and prediction

Here are $N = 10$ points drawn from the function $y(x) = \sin (2 \pi x)$ with some noise added. This is an example of a typical problem you may be faced with, *given the blue dots, what can we infer about the function that generated them, and what might future points look like?*

The data set is composed of pairs of $x$ and $t$ values, $(x_1, t_1)$, $(x_2, t_2)$, ..., $(x_n, t_n)$, 

```julia
d.Image(filename="fig1.2.png") #Bishop Ch. 1
```

To start, let's try fiting the blue dots with a vanilla polynomial, which is of the form

$$y(x, \vec{w}) = w_0 + w_1 x + w_2 x^2 + \ldots + w_M x^M = \sum_{j=0}^M w_j x^j $$

What defines a "good fit" to the data set? One way is to choose an error function and minimize it. A common one is

$$E(\vec{w}) = \frac{1}{2} \sum_{n=1}^N \left [ y(x_n, \vec{w}) - t_n \right]^2 $$

```julia
d.Image(filename="fig1.3.png") #Bishop Ch. 1
```

So, minimizing this function for different values of $M$ gives us mixed results

```julia
d.Image(filename="fig1.4.png") #Bishop Ch. 1
```

So, if we didn't know what the green line looked like beforehand, you might not really know which of these to choose. $M=9$ *does* go through *all* the points, right? Zero error!

## Probability theory

Combine probability theory with decision theory [Sam] and the result is a way to reason about data.

Let $X$ and $Y$ be two random variables. $p(X)$ is the probability that the random variable takes on the value $x$. $p(X,Y)$ is the *joint* probability that $X = x$ and $Y = y$, at the same time.

Sum rule:

$$ p(X) = \sum_Y p(X, Y) $$

We have summed over/integrated out/*marginalized over* $Y$.

Product rule:

$$ p(X, Y) = p(Y| X) p(X) $$

$p(Y | X)$ is a conditional probability, it means what is the probability of $Y$, *given* $X$? 

These two equations form the basis of probability theory. A great reference to learn how to manipulate probability calculus is Hogg 2012 [PDF](http://arxiv.org/abs/1205.4446). 

### Bayes rule

Let's think about our curve fitting example, where we just took the values of the polynomial coefficients $\vec{w}$ that minimized the error. What is the uncertainty in the estimate of these coefficients? 

What we would really like, is a probability function of the form

$$ p(\vec{w} | \mathcal{D})$$

where $\mathcal{D}$ is the dataset that we have, the 10 blue points.

Luckily, we can use *Bayes Theorem* to write down this probability function. This theorem can be derived using the sum and product rules. 

Would anyone like to derive it on the board!?!

$$ p(\vec{w} | \mathcal{D}) = \frac{p(\mathcal{D} | \vec{w}) p(\vec{w})}{p(\mathcal{D})}$$

Ignore the denominator for now.

$$  \overbrace{p(\vec{w} | \mathcal{D})}^{\rm posterior} \propto \underbrace{p(\mathcal{D} | \vec{w})}_{\rm likelihood} \times \overbrace{p(\vec{w})}^{\rm prior}
$$

$$  \overbrace{p( {\rm hypothesis} | {\rm data}, I)}^{\rm posterior} \propto \underbrace{p({\rm data} | {\rm hypothesis},I)}_{\rm likelihood} \times \overbrace{p({\rm hypothesis} | I)}^{\rm prior}
$$

Let's focus on the likelihood function

### Likelihood functions

$$p({\rm data} | {\rm hypothesis},I)$$

Or, alternatively

$$p({\rm data} | {\rm model},I)$$

This is the part of Bayes rule that we can actually do something about. It means given some model of how we expect the world to behave, we can define a probability of observing the data that we have. In a sense, it is a function that we can evaluate, given some model (or model setting).

**NB**: The likelihood is not actually a proper probability distribution over the model ($\vec{w}$) and it is not normalized to 1.

### The Gaussian distribution

$$ \mathcal{N} (x | \mu, \sigma^2) = \frac{1}{(2 \pi \sigma^2)^{1/2}} \exp \left[ -\frac{(x - \mu)^2}{2 \sigma^2} \right] $$

We usually assume that simplest problems have additive Gaussian noise. 

```julia
d.Image(filename="fig1.16.png") #Bishop Ch. 1
```

So, consider a single measurement at ($x_0$, $t_0$). What we mean by this, is that we are assuming that there is a Gaussian "scatter" about $t_0$, defined by the standard deviation of the Gaussian, $\sigma$. Thus, what is the probability we observed a single data point, given our model, i.e., the *likelihood*? It is simply the Gaussian probability distribution

$$p(t_0 | x_0, \vec{w}, \sigma^{-2}) = \mathcal{N}(t_0 | y(x_0, \vec{w}), \sigma^2) $$

Then, assuming that the noise on the data points is independent ("white"/uncorrelated), what is the probability that we observe the full data set? This is the product of the probability of each data point

$$ p(\{ t_0, t_1, \ldots, t_N \} | \{x_0, x_1, \ldots, x_N \}, \vec{w}, \sigma^{-2}) = \prod_{n=1}^N \mathcal{N} \left( t_n | y(x_n, \vec{w}), \sigma^2 \right)$$ 

or, to summarize

$$p({\rm data} | {\rm model},I)$$

## Chapter 2: Probability distributions

**Bernoulli distribution**: binary variables

$$p(x | \mu) = \mu^x (1 - \mu)^{1 - x}$$

* $x$ is the state that the coin is heads
* $\mu$ is the probability of getting heads

**Binomial distribution**: flip the coin $N$ times, how many heads?

**Conjugate prior**: the prior has the same functional form as the likelihood. Means that you now have a closed-form solution for the posterior. The **Beta function** is the conjugate prior for the Binomial distribution.

**Multinomial distribution**: Extension of Bernoulli to more than two states. **Dirichlet distribution** is the conjugate prior.

### Multidimensional Gaussian

There is a reason that the Gaussian distribution keeps cropping up. In multiple dimensions,

$$\mathcal{N}(\mathbf{x} | \mathbf{\mu}, \mathbf{\Sigma}) = \frac{1}{(2 \pi)^{D/2}} \frac{1}{|\mathbf{\Sigma}^{1/2}|} \exp \left [ - \frac{1}{2} (\mathbf{x} - \mathbf{\mu})^T \mathbf{\Sigma}^{-1} (\mathbf{x} - \mathbf{\mu}) \right] $$

Now $\mathbf{\mu}$ is a *vector* and $\mathbf{\Sigma}$ is a covariance *matrix*.
